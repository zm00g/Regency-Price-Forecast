```python
def process_csv_files(directory):
    # Initialize an empty list to store the CSV data
    dataframes = []

    # Iterate over all files in the directory
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)

        # Use magic to identify the file type
        mime = magic.from_file(file_path, mime=True)

        # Check if the file type is application/csv
        if mime == "application/csv":
            with open(file_path, 'r') as f:
                lines = f.readlines()
                data = []
                for line in lines:
                    values = line.strip().split(',')  # Read comma separated line without word wrap
                    data.append(values)
                df = pd.DataFrame(data)
                dataframes.append(df)
    return dataframes

def save_dataframes_to_csv(dataframes, output_filename):
    with open(output_filename, "w") as f:
        for dataframe in dataframes:
            dataframe.to_csv(f, index=False, header=False)

    
```

```python
# Process all CSV files in the specified directory
hoteldataframes = process_csv_files(directory)
```

```python
# Assign output_filename
output_filename = '/Users/zm00g/Data_Science/regency_suites/data/hotel_csv_data.csv'
# Same dataframes to csv
save_dataframes_to_csv(hoteldataframes, output_filename)
```
